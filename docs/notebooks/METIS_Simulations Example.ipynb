# Running METIS_Simulations simulation blocks

This is a notebook example of how to run a simulation block in METIS_Simulations. 

Normally, a simulation block is an executable python file which runs a set of simulations needed to test a recipe or workflow. This notebook does the same thing, but set up interactively, for instructional purposes. 

Note: let the set of simulations run to completion before reading the FITS files into anything, as the last step of the sequence updates some header keyworks. 


```python
# some imports. Set the default IRDB location before importing from metis_simulations

import os
import sys
# change this to wherever you've got the IRDB instance
os.environ['DEFAULT_IRDB_LOCATION']="/Users/me/software/irdb/"
from metis_simulations import runSimulationBlock as rs
```

```python
# some general variables.  If you're running from the command lime the corresponding
# system variables are
# MSIM_YAML_DIR, MSIM_NCORES and MSIM_ODIR

yamlDir = "/Users/me/METIS_Simulations/YAML/"  # where your YAML files are
nCores = 6 # number of cores to use (set to < total number)
outputDir = "./"  # top level output directory
```

```python
# set up the parameters for the simulationBlock

params = {}
# for testing purposes only; no need to change
params['small'] = False  # only changed for testing
params['testRun'] = False # only changed for testing
params['calibFile'] = None # only changed for testing

# These will generally be the same for all runs
params['doStatic'] = True # generate minimal set of static calibration files
params['doCalib'] = 3 # number of each necessary dark/flat to generate 
params['sequence'] = True # generate a sequence of timestamps
params['nCores'] = nCores # home many cores to use

# change for each run
params['outputDir'] = os.path.join(outputDir,"imgLM")  # subdir for output
params['startMJD'] =  "2027-01-25 00:00:00"  # start date for sequence

# list of yamls to be included
yamls = ["scienceLM.yaml","stdLM.yaml","distortionLM.yaml","detlinLM.yaml"]

# get the full path for each YAML
yamlFiles = []
for y in yamls:
    yamlFiles.append(os.path.join(yamlDir,y))
```

```python
# and run. The final empty list in the parameters is where, in command line mode, the command line arguments 
# would be passed. 

rs.runSimulationBlock(yamlFiles,params,[])
```
