We only need to import CPL and the recipe we want (or all of them, with `import *`):

```python
import cpl
import os

from pymetis.instruments.metis.recipes import *
# Convenience functions for loading input framesets
from pymetis.engine.core.functions.frameset import from_filenames, from_tags
```

Then load the desired frameset as a `cpl.ui.FrameSet` from an SOF file (this assumes you have properly set your environment variables `SOF_DATA` and `SOF_DIR`, but you can also specify full paths everywhere):

```python
frameset = cpl.ui.FrameSet(os.path.expandvars("$SOF_DIR/metis_det_dark.ifu.sof"))
```

or from a kwarg mapping `tag: [files]`

```python
frameset = from_tags(DARK_IFU_RAW=['$SOF_DATA/METIS.DARK_IFU_RAW.2027-01-25_00_34_37.fits'],
                     LINEARITY_IFU=['$SOF_DATA/LINEARITY_IFU.fits'],
                     PERSISTENCE_MAP=['$SOF_DATA/PERSISTENCE_MAP.fits'],
                     GAIN_MAP_IFU=['$SOF_DATA/GAIN_MAP_IFU.fits'])
```

or from a dictionary `filename: tag`

```python
frameset = from_filenames({
    '$SOF_DATA/METIS.DARK_IFU_RAW.2027-01-25_00_34_37.fits': 'DARK_IFU_RAW',
    '$SOF_DATA/LINEARITY_IFU.fits': 'LINEARITY_IFU',
    '$SOF_DATA/PERSISTENCE_MAP.fits': 'PERSISTENCE_MAP',
    '$SOF_DATA/GAIN_MAP_IFU.fits': 'GAIN_MAP_IFU'
})
```

and run

```python
MetisDetDark().run(frameset, {})
```
